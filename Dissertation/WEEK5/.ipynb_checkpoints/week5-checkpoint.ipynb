```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np

# CSVGeoDataFrame
csv_file = 'LondonLSOAs.csv'
df = pd.read_csv(csv_file)

# geometryWKT
gdf = gpd.GeoDataFrame(df, geometry=gpd.GeoSeries.from_wkt(df['geometry']))

#  (WGS84)
gdf.set_crs(epsg=4326, inplace=True)

# LSOA
gdf['geo_centroid'] = gdf.centroid

# 
coords = np.array(list(zip(gdf.geo_centroid.x, gdf.geo_centroid.y)))
plt.figure(figsize=(10, 10))
plt.scatter(coords[:, 0], coords[:, 1], s=5, c='blue', alpha=0.6)
plt.title("Centroid Distribution")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()

# 
neighbors = gpd.sjoin(gdf, gdf, how='left', predicate='intersects')

# 
print(neighbors.columns)

# 
neighbors = neighbors[neighbors.index != neighbors.index_right]

# 
G = nx.Graph()

# 
for idx, row in gdf.iterrows():
    G.add_node(idx, pos=(row.geo_centroid.x, row.geo_centroid.y))

# 
for idx, row in neighbors.iterrows():
    G.add_edge(row.name, row['index_right'], weight=1)

# CSV
neighbors_df = neighbors[['LSOA21NM_left', 'LSOA21NM_right']]
neighbors_df.columns = ['node1', 'node2']
neighbors_df.to_csv('neighbors.csv', index=False)

# 
pos = nx.get_node_attributes(G, 'pos')
plt.figure(figsize=(10, 10))
nx.draw(G, pos, node_size=10, node_color='blue', edge_color='grey', alpha=0.5)
plt.title("Adjacency Network")
plt.show()
```

```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
from sklearn.preprocessing import StandardScaler
import torch
from torch_geometric.utils import from_networkx
import seaborn as sns
import torch
import torch.nn.functional as F
from torch_geometric.nn import GATConv
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import random
import itertools

# CSVGeoDataFrame
csv_file = 'LondonLSOAs.csv'
df = pd.read_csv(csv_file)

# geometryWKT
gdf = gpd.GeoDataFrame(df, geometry=gpd.GeoSeries.from_wkt(df['geometry']))

#  (WGS84)
gdf.set_crs(epsg=4326, inplace=True)

#  'c' 
c_features_columns = [col for col in gdf.columns if col.startswith('c')]
c_features = gdf[c_features_columns].apply(pd.to_numeric, errors='coerce')

# 
target_column = 'o_anxiety_quantity_per_capita'
target = gdf[target_column].apply(pd.to_numeric, errors='coerce').fillna(0)

#  'c' 
c_correlation_matrix = c_features.corrwith(target).abs().sort_values(ascending=False)

#  'e' 
e_features_columns = [
    'e_NO2', 'e_ndvi', 'e_water', 'e_trees', 'e_grass', 
    'e_flooded_vegetation', 'e_crops', 'e_shrub_and_scrub', 
    'e_built', 'e_bare', 'e_snow_and_ice'
]
e_features = gdf[e_features_columns].apply(pd.to_numeric, errors='coerce')

#  'e' 
e_correlation_matrix = e_features.corrwith(target).abs().sort_values(ascending=False)

#  'c'  'e' 
all_selected_features = pd.concat([c_features, e_features], axis=1)

# 
for i, row in all_selected_features.iterrows():
    for col in all_selected_features.columns:
        if pd.isna(row[col]):
            neighbor_indices = list(G.neighbors(i))  # NetworkX
            if neighbor_indices:  # 
                neighbor_values = all_selected_features.iloc[neighbor_indices][col]
                mean_value = neighbor_values.mean()
                if pd.isna(mean_value):
                    mean_value = all_selected_features[col].mean()  # nan
                all_selected_features.at[i, col] = mean_value
            else:
                all_selected_features.at[i, col] = all_selected_features[col].mean()  # 

# nan
print("Remaining NaN count per column after filling:\n", all_selected_features.isna().sum())

# 
scaler = StandardScaler()
all_selected_features = scaler.fit_transform(all_selected_features)
all_selected_features = torch.tensor(all_selected_features, dtype=torch.float)

# Tensor
targets = torch.tensor(target.values, dtype=torch.float).unsqueeze(1)  # 

#  NetworkX  PyTorch Geometric 
data = from_networkx(G)
data.x = all_selected_features
data.y = targets  # 

class GAT(torch.nn.Module):
    def __init__(self, in_channels, hidden1, hidden2, out_channels, dropout, heads=1):
        super(GAT, self).__init__()
        self.conv1 = GATConv(in_channels, hidden1, heads=heads, dropout=dropout)
        self.conv2 = GATConv(hidden1 * heads, hidden2, heads=heads, dropout=dropout)
        self.fc1 = torch.nn.Linear(hidden2 * heads, out_channels)
        self.dropout = dropout

    def forward(self, data):
        x, edge_index = data.x, data.edge_index
        x = self.conv1(x, edge_index)
        x = F.relu(x)
        x = F.dropout(x, p=self.dropout, training=self.training)  # Dropout
        x = self.conv2(x, edge_index)
        x = F.relu(x)
        x = F.dropout(x, p=self.dropout, training=self.training)  # Dropout
        x = self.fc1(x)
        return x

def train(model, optimizer, data):
    model.train()
    optimizer.zero_grad()
    out = model(data)
    loss = F.mse_loss(out, data.y)
    loss.backward()
    optimizer.step()
    return loss.item()

# 
param_space = {
    'lr': [0.001, 0.005, 0.01],
    'weight_decay': [5e-4, 1e-3, 5e-3],
    'hidden1': [32, 64, 128],
    'hidden2': [16, 32, 64],
    'dropout': [0.3, 0.5, 0.7],
    'num_epochs': [100, 200]
}

# 
num_samples = 20  # 
random_params = [dict(zip(param_space, v)) for v in random.sample(list(itertools.product(*param_space.values())), num_samples)]

best_loss = float('inf')
best_params = None

for params in random_params:
    model = GAT(
        in_channels=data.x.shape[1], 
        hidden1=params['hidden1'], 
        hidden2=params['hidden2'], 
        out_channels=data.y.shape[1], 
        dropout=params['dropout']
    )
    optimizer = torch.optim.Adam(model.parameters(), lr=params['lr'], weight_decay=params['weight_decay'])

    losses = []
    for epoch in range(params['num_epochs']):
        loss = train(model, optimizer, data)
        losses.append(loss)

    if losses[-1] < best_loss:
        best_loss = losses[-1]
        best_params = params

print(f'Best params: {best_params}, Best loss: {best_loss:.4f}')

# 
model = GAT(
    in_channels=data.x.shape[1], 
    hidden1=best_params['hidden1'], 
    hidden2=best_params['hidden2'], 
    out_channels=data.y.shape[1], 
    dropout=best_params['dropout']
)
optimizer = torch.optim.Adam(model.parameters(), lr=best_params['lr'], weight_decay=best_params['weight_decay'])

losses = []
for epoch in range(best_params['num_epochs']):
    loss = train(model, optimizer, data)
    losses.append(loss)

# 
model.eval()
with torch.no_grad():
    pred = model(data)
    residuals = data.y - pred

# 
y_true = data.y.detach().numpy()
y_pred = pred.detach().numpy()

mse = mean_squared_error(y_true, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_true, y_pred)
r2 = r2_score(y_true, y_pred)

print(f'MSE: {mse:.4f}')
print(f'RMSE: {rmse:.4f}')
print(f'MAE: {mae:.4f}')
print(f'R: {r2:.4f}')

# 
plt.figure(figsize=(10, 5))
plt.hist(residuals.detach().numpy(), bins=50, alpha=0.75)
plt.xlabel('Residual')
plt.ylabel('Frequency')
plt.title('Histogram of Residuals')
plt.show()
```

```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
from sklearn.preprocessing import StandardScaler
import torch
from torch_geometric.utils import from_networkx
import seaborn as sns
import torch.nn.functional as F
from torch_geometric.nn import GATConv
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import random
import itertools

# CSVGeoDataFrame
csv_file = 'LondonLSOAs.csv'
df = pd.read_csv(csv_file)

# geometryWKT
gdf = gpd.GeoDataFrame(df, geometry=gpd.GeoSeries.from_wkt(df['geometry']))

#  (WGS84)
gdf.set_crs(epsg=4326, inplace=True)

#  'c' 
c_features_columns = [col for col in gdf.columns if col.startswith('c')]
c_features = gdf[c_features_columns].apply(pd.to_numeric, errors='coerce')

# 
target_column = 'o_anxiety_quantity_per_capita'
target = gdf[target_column].apply(pd.to_numeric, errors='coerce').fillna(0)

#  'c' 
c_correlation_matrix = c_features.corrwith(target).abs().sort_values(ascending=False)

#  'e' 
e_features_columns = [
    'e_NO2', 'e_ndvi', 'e_water', 'e_trees', 'e_grass', 
    'e_flooded_vegetation', 'e_crops', 'e_shrub_and_scrub', 
    'e_built', 'e_bare', 'e_snow_and_ice'
]
e_features = gdf[e_features_columns].apply(pd.to_numeric, errors='coerce')

#  'e' 
e_correlation_matrix = e_features.corrwith(target).abs().sort_values(ascending=False)

#  'c'  'e' 
all_selected_features = pd.concat([c_features, e_features], axis=1)

# Southwark
is_southwark = gdf['LSOA21NM'].str.contains('Southwark')

# 
train_features = all_selected_features[~is_southwark]
test_features = all_selected_features[is_southwark]
train_target = target[~is_southwark]
test_target = target[is_southwark]

# Graph
train_indices = train_features.index

# NetworkX
train_G = nx.Graph()

# 
for idx, row in gdf.loc[train_indices].iterrows():
    train_G.add_node(idx, pos=(row.geometry.centroid.x, row.geometry.centroid.y))

# 
train_neighbors = gpd.sjoin(gdf.loc[train_indices], gdf.loc[train_indices], how='left', predicate='intersects')

# 
train_neighbors = train_neighbors[train_neighbors.index != train_neighbors.index_right]

# 
for idx, row in train_neighbors.iterrows():
    train_G.add_edge(row.name, row['index_right'], weight=1)

# 
for i in train_indices:
    row = train_features.loc[i]
    for col in train_features.columns:
        if pd.isna(row[col]):
            neighbor_indices = list(train_G.neighbors(i))
            valid_neighbor_indices = [idx for idx in neighbor_indices if idx in train_indices]
            if valid_neighbor_indices:  # 
                neighbor_values = train_features.loc[valid_neighbor_indices, col]
                mean_value = neighbor_values.mean()
                if pd.isna(mean_value):
                    mean_value = train_features[col].mean()  # nan
                train_features.at[i, col] = mean_value
            else:
                train_features.at[i, col] = train_features[col].mean()  # 

# nan
print("Remaining NaN count per column after filling:\n", train_features.isna().sum())

# 
scaler = StandardScaler()
train_features = scaler.fit_transform(train_features)
test_features = scaler.transform(test_features)  # scaler

train_features = torch.tensor(train_features, dtype=torch.float)
test_features = torch.tensor(test_features, dtype=torch.float)

# Tensor
train_targets = torch.tensor(train_target.values, dtype=torch.float).unsqueeze(1)  # 
test_targets = torch.tensor(test_target.values, dtype=torch.float).unsqueeze(1)  # 

#  NetworkX  PyTorch Geometric 
train_data = from_networkx(train_G)
train_data.x = train_features
train_data.y = train_targets  # 

# 
class GAT(torch.nn.Module):
    def __init__(self, in_channels, hidden1, hidden2, out_channels, dropout, heads=1):
        super(GAT, self).__init__()
        self.conv1 = GATConv(in_channels, hidden1, heads=heads, dropout=dropout)
        self.conv2 = GATConv(hidden1 * heads, hidden2, heads=heads, dropout=dropout)
        self.fc1 = torch.nn.Linear(hidden2 * heads, out_channels)
        self.dropout = dropout

    def forward(self, data):
        x, edge_index = data.x, data.edge_index
        x = self.conv1(x, edge_index)
        x = F.relu(x)
        x = F.dropout(x, p=self.dropout, training=self.training)  # Dropout
        x = self.conv2(x, edge_index)
        x = F.relu(x)
        x = F.dropout(x, p=self.dropout, training=self.training)  # Dropout
        x = self.fc1(x)
        return x

def train(model, optimizer, data):
    model.train()
    optimizer.zero_grad()
    out = model(data)
    loss = F.mse_loss(out, data.y)
    loss.backward()
    optimizer.step()
    return loss.item()

# 
param_space = {
    'lr': [0.001, 0.005, 0.01],
    'weight_decay': [5e-4, 1e-3, 5e-3],
    'hidden1': [32, 64, 128],
    'hidden2': [16, 32, 64],
    'dropout': [0.3, 0.5, 0.7],
    'num_epochs': [100, 200]
}

# 
num_samples = 20  # 
random_params = [dict(zip(param_space, v)) for v in random.sample(list(itertools.product(*param_space.values())), num_samples)]

best_loss = float('inf')
best_params = None

for params in random_params:
    model = GAT(
        in_channels=train_data.x.shape[1], 
        hidden1=params['hidden1'], 
        hidden2=params['hidden2'], 
        out_channels=train_data.y.shape[1], 
        dropout=params['dropout']
    )
    optimizer = torch.optim.Adam(model.parameters(), lr=params['lr'], weight_decay=params['weight_decay'])

    losses = []
    for epoch in range(params['num_epochs']):
        loss = train(model, optimizer, train_data)
        losses.append(loss)

    if losses[-1] < best_loss:
        best_loss = losses[-1]
        best_params = params

print(f'Best params: {best_params}, Best loss: {best_loss:.4f}')

# 
model = GAT(
    in_channels=train_data.x.shape[1], 
    hidden1=best_params['hidden1'], 
    hidden2=best_params['hidden2'], 
    out_channels=train_data.y.shape[1], 
    dropout=best_params['dropout']
)
optimizer = torch.optim.Adam(model.parameters(), lr=best_params['lr'], weight_decay=best_params['weight_decay'])

losses = []
for epoch in range(best_params['num_epochs']):
    loss = train(model, optimizer, train_data)
    losses.append(loss)

# 
test_G = nx.Graph()

# 
for idx, row in gdf.loc[is_southwark].iterrows():
    test_G.add_node(idx, pos=(row.geometry.centroid.x, row.geometry.centroid.y))

# 
test_neighbors = gpd.sjoin(gdf.loc[is_southwark], gdf.loc[is_southwark], how='left', predicate='intersects')

# 
test_neighbors = test_neighbors[test_neighbors.index != test_neighbors.index_right]

# 
for idx, row in test_neighbors.iterrows():
    test_G.add_edge(row.name, row['index_right'], weight=1)

#  NetworkX  PyTorch Geometric 
test_data = from_networkx(test_G)
test_data.x = test_features
test_data.y = test_targets  # 

# Southwark
model.eval()
with torch.no_grad():
    pred = model(test_data)
    pred = pred[:len(test_targets)]  # 
    residuals = test_data.y - pred

# 
y_true = test_data.y.detach().numpy()
y_pred = pred.detach().numpy()

# NaN
if np.isnan(y_pred).any():
    print("NaN values found in predictions")

# NaN
y_true = y_true[~np.isnan(y_pred)]
y_pred = y_pred[~np.isnan(y_pred)]

mse = mean_squared_error(y_true, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_true, y_pred)
r2 = r2_score(y_true, y_pred)

print(f'MSE: {mse:.4f}')
print(f'RMSE: {rmse:.4f}')
print(f'MAE: {mae:.4f}')
print(f'R: {r2:.4f}')

# 
plt.figure(figsize=(10, 5))
plt.hist(residuals.detach().numpy(), bins=50, alpha=0.75)
plt.xlabel('Residual')
plt.ylabel('Frequency')
plt.title('Histogram of Residuals')
plt.show()
```
